```python
print( "--- Finding RSA fixed points for p, q, e ---")

#Change p,q,e
p=71
q=41
e = 11
######
n=p*q;
phi=(p-1)*(q-1)
gcd_e=gcd(e, phi)


print( "Prime p: ",p);
print ("Prime q: ",q);
print ("Modul n: ",n);
print( "Public exponent e: ", e, gcd_e==1)


R=Integers(p)
gen_f_p = R.multiplicative_generator();
print( "Generator of f_p: ",gen_f_p);
S=Integers(q)
gen_f_q = S.multiplicative_generator();
print ("Generator of f_q: ",gen_f_q);

#calculating number of fixed points
gcd_p = gcd(e-1,p-1)
gcd_q = gcd(e-1,q-1)
print ("gcd(e-1,p-1): ", gcd_p);
print ("gcd(e-1,q-1): ", gcd_q);
NP=(gcd_p+1)*(gcd_q+1)
print ("Number of fixpoints: ",NP)

#primitive roots mod p and mod q
#print ("Fixpoints modulo p");
fp=[0]
fq=[0]
wp=power_mod(gen_f_p,Integer((p-1)/gcd_p),p);
#print('wp',wp)


wq=power_mod(gen_f_q,Integer((q-1)/gcd_q),q);
#print('wq',wq)

for i in range(gcd_p):
    fix_p = power_mod(wp,i,p);
    fp.append(fix_p)
#print (fp)
#print( "Fixpoints modulo q");
for j in range(gcd_q):
    fix_q = power_mod(wq,j,q);
    fq.append(fix_q);
#print( fq)

# Calculating fixed points using chinese remainder theorem
Ip=CRT([1,0],[p,q])
Iq=CRT(0,1,p,q)
print("Ip=",Ip,"Iq=",Iq)
print ("\nFixpoints for the public RSA key (n,e) = (", n, ",", e, ")")
print('================================')
L=[]
for r in fp:
    for s in fq:
        F = int((Ip*Integer(r)+Iq*Integer(s))%n)
        L.append(F)
        #print(r,s,F,gcd(F,n),gcd(F+1,n),gcd(F-1,n))
print(L)

########################################################
# fixed point F is not usable for RSA factorisation if ord(F) mod p= ord(F) mod q
dd=gcd([p-1,q-1,e-1]);
Mdiv=divisors(dd)
S=1
for r in Mdiv:
    S=S+euler_phi(r)^2
print("number of fixed points not usable for factorization",S)
```
